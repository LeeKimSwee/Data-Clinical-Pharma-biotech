```python
os.chdir('/Users/kim/Desktop/AllPublicXML/NCT0000xxxx')
df = pd.DataFrame()
for file in os.listdir()[0:10]:
    tree = ET.parse(file)
    root = tree.getroot()
    features = {}
    for child in root:
        if len(child.getchildren()) == 0:
            features[child.tag] = [child.text]
        else:
            for elem in child:
                # since mesh term appears several times, need to append
                # first check if it exists
                if child.tag + '_' + elem.tag in features:
                    # code to append
                    features.setdefault(str(child.tag + '_' + elem.tag), []).append(elem.text)                    
                else:
                    features[child.tag + '_' + elem.tag] = [elem.text]
    df = df.append(features, ignore_index=True)
df.head(20)
```

```python
for serie in list(df):
    df[serie] = df[serie].astype(str)
```

```python
def brackets_eraser(text):
    bracketsless_text = text.str.extract(r'\[(.*)\]')
    return bracketsless_text   
```

```python
def newline_eraser(text):
    text_without_new_line_and_space = text.str.replace(r'\\n\s+', '')
    return text_without_new_line_and_space
```
