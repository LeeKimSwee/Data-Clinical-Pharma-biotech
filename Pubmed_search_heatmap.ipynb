```python
import urllib.parse
import urllib.request
import re
```

```python
searchlist = ['MDSC','Ploegh hl','Swee lk']
url = 'https://www.ncbi.nlm.nih.gov/pubmed/?'
hits = []
for i in searchlist:
    data = {'term':str(i)}
    values = urllib.parse.urlencode(data)
    req = url + values
    resp = urllib.request.urlopen(req)
    x = resp.read()
    text = x.decode('utf-8')
    number = re.search(r'id="resultcount" value="(.*)" /><input name', text)
    hits.append(number.group(1))
print (hits)
```
